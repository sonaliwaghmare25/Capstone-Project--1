# Join SQL Queries

```python
df1=pd.read_csv('Table1.csv')
df1
```

```python
df2=pd.read_csv('Table3.csv')
df2
```

```python
df3=pd.read_csv('Table2.csv')
df3
```

```python
df=pd.concat([df1,df2,df3])
df
```

```python
df.isnull().sum()
```

```python
df['Colors']=df[['Colors']].fillna('5 colors')
```

```python
df['Price']=df[['Price']].fillna('₹7 999.00')
```

```python
df.drop('Unnamed: 0.1',axis=1,inplace=True)
```

```python
df
```

```python
df['Shoename']=df[['Shoename']].fillna('Woodland Trekking Shoes')
```

```python
df['Reviews']=df[['Reviews']].fillna('Reviews(100)')
```

```python
df['Size ']=df[['Size']].fillna('36')
```

```python
df['Comfort']=df[['Comfort']].fillna('Outstanding')
```

```python
df['Quality']=df[['Quality']].fillna('Excellant')
```

```python
df['Star_ratings']=df[['Star_ratings']].fillna('4')
```

```python
df.set_index('Shoename',inplace=True)
```

```python
df['Size_count']=df[['Size_count']].fillna('6')
```

```python
df['Color1']=df[['Color1']].fillna('GRAY')
```

```python
df['Color2']=df[['Color2']].fillna('BLACK/WHITE')
```

```python
df['Style_code']=df[['Style_code']].fillna('120473')
```

```python
df['Color3']=df[['Color3']].fillna('OLIVE GREEN/RED')
```

```python
df['Color4']=df[['Color4']].fillna('CAMEL')
```

```python
df['Color5']=df[['Color5']].fillna('Vintage Pink, Nimbus Grey')
```

```python
df['Size']=df[['Size']].fillna('39')
```

```python
df['Category']=df[['Category']].fillna('Unisex')
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```
