```python
from scipy.stats import zscore
outliers=df[['Shoename', 'Category', 'Colors', 'Price', 'Reviews', 'Size', 'Comfort',
       'Quality', 'Star_ratings', 'Size_count', 'Color1', 'Color2',
       'Style_code', 'Color3', 'Color4', 'Color5']]
z=np.abs(zscore(outliers))
z
```

The above table shows the zscore values for all features.

```python
np.where(z>3)
```

```python
z.iloc[99,4]
```

```python
df1=df[(z<3).all(axis=1)]
df1
```

```python
print('Old data shape',df.shape)
print('New data shape',df1.shape)
```

```python
print('Data loss percentage',((df.shape[0]-df1.shape[0])/df.shape[0])*100)
```

Now we will try to check and remove outliers from IQR method.

```python
Q1=df1.quantile(.25)
Q3=df1.quantile(.75)
IQR=Q3-Q1
IQR
```

```python
df1_new=df1[(z<3).all(axis=1)]
df1_new
```

Both the zscore and QR methods are showing the same result for outliers removal. Hence, we have cleared the outliers from the data.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
pred_lr=lr.predict(x_test)
pred_lr
```

```python
print('r2 score:',r2_score(y_test,pred_lr))
print('mean_squared_error:', mean_squared_error(y_test,pred_lr))
print('root mean_squared_error:',np.sqrt(mean_squared_error(y_test,pred_lr)))
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor()
rfr.fit(x_train,y_train)
pred_rfr=rfr.predict(x_test)
pred_rfr
```

```python
print('r2 score:',r2_score(y_test,pred_rfr))
print('mean_squared_error:', mean_squared_error(y_test,pred_rfr))
print('root mean_squared_error:',np.sqrt(mean_squared_error(y_test,pred_rfr)))
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor()
knn.fit(x_train,y_train)
pred_knn=knn.predict(x_test)
pred_knn
```

```python
print('r2 score:',r2_score(y_test,pred_knn))
print('mean_squared_error:', mean_squared_error(y_test,pred_knn))
print('root mean_squared_error:',np.sqrt(mean_squared_error(y_test,pred_knn)))
```

```python
from sklearn.linear_model import Lasso
ls=Lasso()
ls.fit(x_train,y_train)
pred_ls=ls.predict(x_test)
pred_ls
```

```python
print('r2 score:',r2_score(y_test,pred_ls))
print('mean_squared_error:', mean_squared_error(y_test,pred_ls))
print('root mean_squared_error:',np.sqrt(mean_squared_error(y_test,pred_ls)))
```

```python
from sklearn.linear_model import Ridge
rdg=Ridge()
rdg.fit(x_train,y_train)
pred_rdg=rdg.predict(x_test)
pred_rdg
```

```python
print('r2 score:',r2_score(y_test,pred_rdg))
print('mean_squared_error:', mean_squared_error(y_test,pred_rdg))
print('root mean_squared_error:',np.sqrt(mean_squared_error(y_test,pred_rdg)))
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
pred_dtr=dtr.predict(x_test)
pred_dtr
```

```python
print('r2 score:',r2_score(y_test,pred_dtr))
print('mean_squared_error:', mean_squared_error(y_test,pred_dtr))
print('root mean_squared_error:',np.sqrt(mean_squared_error(y_test,pred_dtr)))
```

```python
from sklearn.ensemble import ExtraTreesRegressor
etr=ExtraTreesRegressor()
etr.fit(x_train,y_train)
pred_etr=etr.predict(x_test)
pred_etr
```

```python
print('r2 score:',r2_score(y_test,pred_etr))
print('mean_squared_error:', mean_squared_error(y_test,pred_etr))
print('root mean_squared_error:',np.sqrt(mean_squared_error(y_test,pred_etr)))
```

From the above model fitting and performance score, we can say that, RandomForestRegressor is the best fitting and best performing models,which is giving a performance score as 97.50.
